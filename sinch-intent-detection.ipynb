# Intent Detection

> Importing the libraries

```python
import pandas as pd
```

```python
sample = pd.read_csv('/kaggle/input/intent-detection-by-sinch/sample.csv')
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import StandardScaler

# Load data
X_train = np.load('/kaggle/input/intent-detection-by-sinch/X_train.npy')
X_test = np.load('/kaggle/input/intent-detection-by-sinch/X_test.npy')
y_train = pd.read_csv('/kaggle/input/intent-detection-by-sinch/y_train.csv').Predicted.to_numpy()

# Normalize the input data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Remove extra dimension from X_train and X_test
X_train = np.squeeze(X_train)
X_test = np.squeeze(X_test)

# Define the model
model = Sequential()
model.add(Dense(units=512, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.2))
model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=45, activation='softmax'))

# Compile model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
model.fit(X_train, y_train, validation_split=0.2, epochs=50, batch_size=64)
```

```python
# Make predictions on test set
y_pred = model.predict(X_test)

# Create submission DataFrame
submission = pd.DataFrame({'id': sample['Id'], 'Predicted': y_pred.ravel()[:len(sample)]})

# Save submission to CSV file
submission.to_csv('submission.csv', index=False)
```
