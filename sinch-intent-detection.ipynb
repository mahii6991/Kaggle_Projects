# using the deep learning

```python
#from here we will start by defining the model.
import keras.backend as K
from keras.layers import Dense , Embedding , LSTM ,SimpleRNN
from keras.models import Sequential
```

```python
embedding_layer = Embedding(200, 45, input_length=50)
```

```python
print(f"the {X_train_new.shape} shape of the X_train")
print(f"the {y_train_new.shape} shape of the y_train")
```

```python
#finding out the shape of the x_train
X_test_new = pd.DataFrame(X_test)
#printing out the results
print(f"the {X_test_new.shape} shape of the X_test")
```

```python
model = Sequential()

model.add(embedding_layer)
model.add(LSTM(32,dropout=0.2,recurrent_dropout=0.2))
model.add(Dense(32,activation='relu'))
#model.add(SimpleRNN(32, return_sequences=False))
#model.add(Dense(32,activation='tanh'))
model.add(Dense(45,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X_train_new,y_train_new,batch_size=32,epochs=10)
```
