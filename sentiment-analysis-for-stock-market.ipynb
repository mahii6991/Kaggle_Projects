```python
from transformers import pipeline#importing pipeline from the transformers
```

```python
classifier = pipeline("zero-shot-classification", device=0)
```

```python
headline_text_class = news1_copy['headline_text']
```

```python
headline_text_class.dtypes
```

```python
#need to convert it into the list form
headline_list = headline_text_class.values.tolist()
```

```python
headline_list[0:5]
```

```python
#we have ran the model on the 10000 of the tweets other wise it would have been 
sequence = headline_list[0:10000]
```

```python
candidate_labels = ["postive","negative"]
```

```python
classified_text = classifier(sequence,candidate_labels,multi_label=False)
```
