# Starting off with the mickey code

```python
import nltk
import re
from nltk.tokenize import word_tokenize
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
#from keras.preprocessing.text import Tokenizer
#from keras_preprocessing.sequence import pad_sequences
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multiclass import OneVsRestClassifier
from scipy import sparse
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')


print("All packages are load")
```

```python
econbiz['title'] = econbiz['title'].map(lambda x : x.lower())
econbiz.head()
```

# stopwords

```python
from nltk.corpus import stopwords
```

```python
english_stop_word = set(stopwords.words('english'))
econbiz['title'] = econbiz['title'].apply(lambda x: ' '.join([word for word in x.split() if word not in (english_stop_word)]))
```

```python
econbiz.head()
```

# Lemmatization

```python
from textblob import TextBlob, Word, Blobber
from textblob.classifiers import NaiveBayesClassifier
from textblob.taggers import NLTKTagger
```

```python
econbiz['title'] = econbiz['title'].apply(lambda x: " ". join([w.lemmatize() for w in TextBlob(x).words]))
```

```python
econbiz.head()
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stem_english = SnowballStemmer("english")
econbiz['title'] = econbiz['title'].apply(lambda x: stem_english.stem(x)) 
```

```python
econbiz.head()
```

```python
econbiz['title'] = econbiz['title'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x)) #Removes numbers
econbiz['title'] = econbiz['title'].apply(lambda x: re.sub('\s+[a-zA-Z]\s+', ' ', x)) # Removes single characters
econbiz['title'] = econbiz['title'].apply(lambda x: re.sub('\s+', ' ', x))
```

# Tokenizing 

```python
econbiz['title'] = econbiz['title'].apply(lambda x: word_tokenize(x))
```

```python
econbiz.head()
```

# Cleaning Labels

```python
econbiz['labels'] = econbiz['labels'].map(lambda x : x.split('\t'))
```

```python
econbiz.head()
```

# Joining the Labels

```python
econbiz['title'] = econbiz['title'].map(lambda x : ' '.join(x))
econbiz['labels'] = econbiz['labels'].map(lambda x : ' '.join(x))
```

```python
econbiz.head()
```

```python
econbiz = econbiz[20000:]
```

```python
X = econbiz['title']
```

```python
X.head()
```

```python
y = econbiz['labels']
```

```python
y.head()
```

# Encoding the y_label

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
```

```python
y_trans = mlb.fit_transform(y)
```

# splitting the data set 

```python
X.shape
```

```python
np.shape(y_trans)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_trans, test_size=0.4, random_state=23)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2')
x_train = vectorizer.fit_transform(X_train)
x_test = vectorizer.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
```

```python
type(y_train)
```

```python
y_train
```

```python
from scipy import sparse
```

```python
sy_train = sparse.csr_matrix(y_train) 
```

```python
sy_test = sparse.csr_matrix(y_test) 
```

```python
# using binary relevance
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB

# initialize binary relevance multi-label classifier
# with a gaussian naive bayes base classifier
classifier = BinaryRelevance(GaussianNB())

# train
classifier.fit(x_train, sy_train)

# predict
predictions = classifier.predict(x_test)

# accuracy
print("Accuracy = ",accuracy_score(sy_test,predictions))
print("\n")
```

```python
# using Label Powerset
from skmultilearn.problem_transform import LabelPowerset
# initialize label powerset multi-label classifier
classifier = LabelPowerset(LogisticRegression())
# train
classifier.fit(X_train, sy_train)
# predict
predictions = classifier.predict(sy_test)
# accuracy
print("Accuracy = ",accuracy_score(sy_test,predictions))
print("\n")
```
