# Starting out with the modelling of data

```python
data[60]=data[60].replace('R',0)
```

```python
data[60]= data[60].replace('M',1)
```

```python
X = data.drop(60,axis="columns") 
y = data[60]
```

```python
X.head()
```

```python
y.head()#we need to change the object type of the y predictor 
```

```python
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeClassifier
from sklearn.preprocessing import StandardScaler


X_train,X_test,y_train,y_test = train_test_split(X , y, test_size=0.30, random_state=36)
```

```python
i = [X_train,X_test,y_train,y_test]

for c in i:
    print(Counter(c))
```

# RidgeClassifier

It first converts the dataset into the rango {-1,1}, and then treats the problem as the regression.

```python
ridge = RidgeClassifier()
ridge.fit(X_train,y_train)
```

```python
#finding out the score on the training dataset
ridge.score(X_train,y_train)
```

```python
from sklearn import metrics
#finding out the accuracy on the test dataset
#making out the prediction
yhat = ridge.predict(X_test)
#accuracy
accuracy = metrics.accuracy_score(y_test,yhat)
print("The accuracy score when using the RidgeClassifier is {}.".format(accuracy*100))
```

# MultiLayer Perceptron

```python
from sklearn.neural_network import MLPClassifier

small_neural = MLPClassifier(hidden_layer_sizes=(5, 2),activation='relu',solver="adam",random_state=1)
```

```python
small_neural.fit(X_train,y_train)
```

```python
small_neural.score(X_train,y_train)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logit= LogisticRegression() #we the defualt solver in this model is 'lbfgs'
logit.fit(X_train,y_train)
```

```python
yhat = logit.predict(X_test)
accuracy = metrics.accuracy_score(y_test,yhat)

print("The accuracy when using the logistic regression model is {} %.".format(round(accuracy,2)*100))
```
